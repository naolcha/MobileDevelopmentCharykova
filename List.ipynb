# Списки(List)

| Массив | Список |
| - | - |
| статистический размер | динам размер |
| быстрей | удобней |

```
// Не типизированный список (Необобщённый список)

using System.Collections;

ArrayList arrayList = new();
arrayList.Add(3);
arrayList.Add("Hey");
arrayList.Add(3.14);
arrayList
```

```
// Типизированный список (Обобщённый список)

List<int> numbers = new();

numbers.Add(20);
numbers.Add(1);
numbers.Add(-231);
numbers.Add(341);
numbers
```
