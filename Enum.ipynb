# Перечисления (enum)

```
enum TaskStatus {Todo, InProgress, Done, Close};
TaskStatus taskStatus;
Console.WriteLine(taskStatus); // <-- по умолчанию
taskStatus = TaskStatus.InProgress;
Console.WriteLine(taskStatus);
Console.WriteLine(taskStatus.ToString("D"));

// if (taskStatus == 1) ...
// if (taskStatus == TaskStatus.InProgress) ...
```

```
enum MultiColor: byte
{
    Red = 1, // 0b_0000_0001
    Green = 2, // 0b_0000_0010
    Blue = 4, // 0b_0000_0100
    Black = 8,  // 0b_0000_1000
    RedBlue = Red | Blue
} 
MultiColor multiColo;
Console.WriteLine(multiColo);
multiColo = MultiColor.Red;
multiColo |= MultiColor.Blue;
Console.WriteLine(multiColo);
multiColo ^= MultiColor.Blue;
Console.WriteLine(multiColo);
```

```
multiColo = MultiColor.RedBlue;
// Проверки
Console.WriteLine($"Это только Red = {multiColo == MultiColor.Red}");
Console.WriteLine($"Это только Red = {multiColo.HasFlag(MultiColor.Red)}");
```

```
for (int i = 0; i<8;i++)
{
    var r = 1 << i;
    Console.WriteLine($"1 << {i} = {r, 3} || {r, 8:B8}");
}
```

```
byte n = 0b_0000_0111;
byte numBit = 2;
Console.WriteLine($"{1 << numBit:B8}");
Console.WriteLine((n & 1 << numBit) != 0);
```
