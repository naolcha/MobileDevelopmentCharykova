# Генератор функция (метод)

```
IEnumerator GetCities()
{
    yield return "Калуга";
    yield return "Воронеж";
    yield return "Тула";
}

var x = GetCities();
Console.WriteLine(x.Current);
Console.WriteLine(x.MoveNext());
Console.WriteLine(x.Current);
```

```
IEnumerable<string> GetCities()
{
    yield return "Калуга";
    yield return "Воронеж";
    yield return "Тула";
}

foreach (var i in GetCities())
    Console.WriteLine(i);

```

```
IEnumerable<int> GetNumbers(int start, int end)
{
    for (int i = start; i < end; i++)
        yield return i;
}

foreach (var i in GetNumbers(10, 15))
    Console.WriteLine(i);
```
