# LINQ

```
var numbers = Enumerable.Range(0, 10).ToArray();
Console.WriteLine(string.Join(", ", numbers));

var q_numbers = 
    from v in numbers
    where v > 5 && v < 8
    orderby v descending
    select $"__{v}__";
Console.WriteLine(string.Join(", ", q_numbers));
```

```
var cites = new string[] {"Москва", "Сочи", "Самара", "Ялта", "Суммы"};
var q1 = cites.Where(i => i.EndsWith('а')).OrderBy(i => i).ToArray();
Console.WriteLine(string.Join(", ", q1));
```
