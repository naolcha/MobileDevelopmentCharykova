# Аргументы функции (методов)

- по значению
- по ссылке

```
// Передача параметров по значению

void Func1(int n)
{
    n += 100;
    Console.WriteLine(n);
}
int x = 1;
Func1(x);
x
```

```
// Передача параметров по ссылке

void Func2(ref int n)
{
    n += 100;
    Console.WriteLine(n);
}
int x = 1;
Func2(ref x);
x
```

```
// Передача параметров по ссылке только для чтения

void Func3(ref readonly int n)
{
    //n += 100;
    Console.WriteLine(n);
}
int x = 1;
Func3(ref x);
x
```

```
// Передача параметров по out

void Func4(out int n)
{
    // n += 100; // нельзя
    n = 50;
}
int x = 1;
Func4(out x);
x
```

```
Func4(out int y);
y
```

```
// Передача параметров по in

void Func5(in int n)
{
    // n += 100; // нельзя
    Console.WriteLine(n);
}
int x = 1;
Func5(in x);
x
```
