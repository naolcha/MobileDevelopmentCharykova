# Работа с текстовыми файлами

```
using System.IO;
```

```
const string TEMP_DIR = "temp";

if (!Directory.Exists(TEMP_DIR))
{
    Directory.CreateDirectory(TEMP_DIR);
}
```

```
Path.Combine(TEMP_DIR, "aaa", "bbb", "temp_TXT.txt")
```

```
Path.Combine("C:", "aaa", "bbb", "temp_TXT.txt")
```

```
var  fileName = Path.Combine(TEMP_DIR, "temp_TXT.txt");
Console.WriteLine(fileName);
// запись в файл

var f = new StreamWriter(fileName);
try
{
    f.Write("Hello");
    f.WriteLine("123");
    f.WriteLine("Привет");
}
finally
{
    f.Close();
}
```

```
using (var f = new StreamWriter(fileName))
{
    f.Write("Hello");
    f.WriteLine("123");
    f.WriteLine("Привет");
}
```

```
{
    using var f = new StreamWriter(fileName);
    f.Write("1");
    f.WriteLine("123");
    f.WriteLine("Привет");
}
```

```
// Добавить в конец файла
using (var f = new StreamWriter(fileName, append:true))
{
    f.Write("Hello");
    f.WriteLine("123");
    f.WriteLine("Привет");
}
```

```
// Чтение из файла (целтликом)
using (var f = new StreamReader(fileName))
{
    var lines = f.ReadToEnd();
    Console.WriteLine(lines);
}
```

```
// Чтение из файла (по строчно)
using (var f = new StreamReader(fileName))
{
    string lines;
    while (f.EndOfStream == false)
    {
        lines = f.ReadLine();
        Console.WriteLine(lines);
    }
}
```

```
// ... или
using (var f = new StreamReader(fileName))
{
    string lines;
    while ((lines = f.ReadLine()) != null)
    {
        Console.WriteLine(lines);
    }
}
```

```
var  fileName1 = Path.Combine(TEMP_DIR, "temp_TXT2.txt");

// Перезаписать файл
File.WriteAllText(fileName1, "aaa\ndasdas\n12");
// Дозаписать в файл
File.AppendAllText(fileName1, DateTime.Now.ToString());
// Чтение из файла в строку 
var st = File.ReadAllText(fileName1);
Console.WriteLine(st);
// Чтение из файла в массив
var lines = File.ReadAllLines(fileName1);
lines
```
