# Работа с текстовыми файлами

```
using System.IO;

const string TEMP_DIR = "temp"; //константы пишутся капсом с нижним подчёркиванием
if (! Directory.Exists(TEMP_DIR))
    Directory.CreateDirectory(TEMP_DIR);

//TEMP_DIR + "\\" + "temp_TXT.txt"
var filename = Path.Combine(TEMP_DIR, "temp_TXT.txt");
Console.WriteLine(filename);

var f = new StreamWriter(filename);
try
{
    f.Write("Hello");
    f.WriteLine("123");
    f.WriteLine("AAA");
}
finally
{
    f.Close();
}

```

```
using (var f = new StreamWriter(filename))
{
    f.Write("Hello");
    f.WriteLine("123");
    f.WriteLine("AAA");
}
```

```
// или
{
    using var f = new StreamWriter(filename);
    f.Write("Hello");
    f.WriteLine("123");
    f.WriteLine("AAA");
}
```

```
//добавить в конец
using (var f = new StreamWriter(filename, append:true))
{
    f.Write("Hello");
    f.WriteLine("123");
    f.WriteLine("Привет");
}
```

```
//Чтение из файла
using (var f = new StreamReader(filename))
{
    var lines = f.ReadToEnd;
    Console.WriteLine(lines);
}
```

```
using (var f = new StreamReader(filename))
{
    string line;
    while (f.EndOfStream == false)
    {
        var lines = f.ReadToEnd;
        Console.WriteLine(lines);
    }
}
```
