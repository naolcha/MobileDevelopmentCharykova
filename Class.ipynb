# Класс(class)

```
// (1) Класс с полями и методами

class Person
{
    // поля
    public string name = "";
    public int age = 18; // <-- по умолчанию
    // Методы
    public string Say() => "Hello!!";
}

var p = new Person();
Console.WriteLine($"{p.name} {p.age}");

// Через поля
p.name = "Ivan";
p.age = 190;
Console.WriteLine($"{p.name} {p.age}");

// Через иницилизатор
var p2 = new Person(){name = "KOLO", age = 90};
Console.WriteLine($"{p2.name} {p2.age}");
```

```
// (2) Гетеры и Сетторы

class Person
{
    // поля
    private string name = "";
    private int age = 18;
    // Методы Геттеры и Сетторы
    public string GetName() => name;
    public void SetName(string value) => name = value;
    public int GetAge() => age;
    public void SetAge(int value) => age = value;
}
var p4 = new Person();
p4.SetName("POLL");
p4.SetAge(19);
Console.WriteLine($"{p4.GetName()} {p4.GetAge()}");
```

```
// (3) Свойства
class Person
{
    // поля
    private string name = "";
    private int age = 18;
    // Свойства
    public string Name
    {
        get
        {
            return name;
        }
        set
        {
            name = value;
        }
    }
    public int Age {get => age; set => age = value;}

}


Person p3 = new Person();
p3.Age = 12;
p3.Name = "daw";
Console.WriteLine($"{p3.Name} {p3.Age}");
```

```
// оператор init
// МОЖНО (только) установить
// - либо в объявлениях
// - либо в конструкторах
// - либо в иницилизаторае

class Person()
{
    public string Name {get; init; }
    public int Age {get; set; }
}
var p = new Person() {Name = "Миша"};
// p.na,e = "Ivan" нельзя
```

```
// Модификатор required
// Нужно (обязательно) установить
// - либо в конструкторе
// - либо в инициаторе
class Person()
{
    public required string Name {get; set; }
    public required int Age {get; set; }
}
var p = new Person() {Name = "Миша", Age = 23};
```

```
// Декострукторы (не путать с декструктором)
class Person(string name, int age)
{
    public string Name {get; set; } = name;
    public int Age {get; set; } = age;
    public void Deconstruct(out string name, out int age)
    {
        name = Name;
        age = Age;
    }
    public void Deconstruct(out string name, out int age, out string text)
    {
        name = Name;
        age = Age;
        text = "FJO";
    }
}
var p = new Person("Yara", 19);
(string name, int age) = p;
Console.WriteLine(name);
Console.WriteLine(age);

(_, _, string text) = p;
Console.WriteLine(text);
```

```
// Наследование
class Person()
{
    public string Name {get; init; }
}
class Employee : Person
{
    public string Company {get; set; }
}

var e = new Employee() {Name = "Ivad", Company = "Yandex"};
```

```
// Наследование + Конструктор
class Person()
{
    public string Name {get; init; }
    public Person(string name)
    {
        Name = name;
    }
}
class Employee : Person
{
    public string Company {get; set; }
    public Employee(string name, string company) : base(name)
    {
        Company = company;
    }
}

var e = new Employee(name:"Icad", company:"dawd");
```

```
// Индексаторы

class Code(string text)
{
    public string Text {get; } = text;
    public int Lenght{ get => Text.Length;}
    public string this[Index index]
    {
        get => new string(Text[index], 5);
    }
}
var code = new Code("Heollo");
Console.WriteLine(code[^1]);

for (int i = 0; i < code.Lenght; i++)
    System.Console.WriteLine($"code[{i}] = {code[i]}");

// Ошибка потому что нет GetEnumerator
// foreach (var i in code)
//     Console.WriteLine(i);
```

```
// ... + GetEnumerator

class Code(string text)
{
    public string Text {get; } = text;
    public int Lenght{ get => Text.Length;}
    public string this[Index index]
    {
        get => new string(Text[index], 5);
    }
    public IEnumerator GetEnumerator() => Text.GetEnumerator();
}

var code = new Code("Heollo");
foreach (var i in code)
    Console.WriteLine(i);

// Console.WriteLine(string.Join(", ", code.Text));
```
