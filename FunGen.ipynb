# Генератор функция

```
IEnumerator GetCites()
{
    yield return "Калуга";
    yield return "Воронеж";
    yield return "Тула";
    yield return "Москва";
}

var x = GetCites();
Console.WriteLine(x.Current);
Console.WriteLine(x.MoveNext());
Console.WriteLine(x.Current);
```

```
IEnumerable<string> GetCites()
{
    yield return "Калуга";
    yield return "Воронеж";
    yield return "Тула";
    yield return "Москва";
}

foreach (var i in GetCites())
{
    Console.WriteLine(i);
}
```

```
IEnumerable<int> GetNumbers(int start, int end)
{
    for (int i = start; i < end; i++)
    {
        yield return i;
    }
    // yield return "Калуга";
    // yield return "Воронеж";
    // yield return "Тула";
    // yield return "Москва";
}

foreach (var i in GetNumbers(17, 20))
{
    Console.WriteLine(i);
}
```
