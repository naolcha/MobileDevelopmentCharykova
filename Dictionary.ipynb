# Словарь (Dictionary)

```
Dictionary<int, string> d = new();
d.Add(4, "Good");
d.Add(5, "Great");
d.Add(3, "OK");
// добавть
d[2] = "bad";
// редактировать
d[5] = "EXELENT!";
// удалить
d.Remove(4);
d
```

```
foreach (var i in d)
Console.WriteLine($"{i.Key} | {i.Value}");
```

```
foreach (var(key, value) in d)
Console.WriteLine($"{key} | {value}");
```

```
string value;
d.TryGetValue(4, out value);
value
```

```
d.TryGetValue(4, out string value2);
value2
```

```
// проверить наличия ключа или значения
Console.WriteLine(d.ContainsKey(4));
Console.WriteLine(d.ContainsKey(5));
Console.WriteLine(d.ContainsValue("bad"));
```
