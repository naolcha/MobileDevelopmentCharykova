# Печать фмгур (ДЗ)

```
public enum SnakeDirection
{
    Clockwise,
    CounterClockwise
}

public enum SnakeOrientation
{
    Horizontal,
    Vertical
}


     void PrintFigure(string[] figure)
    {
        foreach (string line in figure)
        {
            Console.WriteLine(line);
        }
    }

    static char[,] CreateCanvas(int width, int height, char backgroundChar = ' ')
    {
        char[,] canvas = new char[height, width];

        for (int y = 0; y < height; y++)
        {
            for (int x = 0; x < width; x++)
            {
                canvas[y, x] = backgroundChar;
            }
        }

        return canvas;
    }

    static string[] ToLines(char[,] canvas)
    {
        int height = canvas.GetLength(0);
        int width = canvas.GetLength(1);
        string[] lines = new string[height];

        for (int y = 0; y < height; y++)
        {
            char[] row = new char[width];
            for (int x = 0; x < width; x++)
            {
                row[x] = canvas[y, x];
            }
            lines[y] = new string(row);
        }

        return lines;
    }

    static string[] TransposeFigure(string[] figure)
    {
        if (figure.Length == 0)
        {
            return Array.Empty<string>();
        }

        int height = figure.Length;
        int width = figure[0].Length;
        string[] result = new string[width];

        for (int x = 0; x < width; x++)
        {
            char[] row = new char[height];
            for (int y = 0; y < height; y++)
            {
                row[y] = figure[y][x];
            }
            result[x] = new string(row);
        }

        return result;
    }

    static void DrawRectangleBorder(char[,] canvas, int left, int top, int width, int height, char borderChar, int borderThickness = 1)
    {
        for (int layer = 0; layer < borderThickness; layer++)
        {
            int currentLeft = left + layer;
            int currentTop = top + layer;
            int currentRight = left + width - 1 - layer;
            int currentBottom = top + height - 1 - layer;

            if (currentLeft > currentRight || currentTop > currentBottom)
            {
                break;
            }

            for (int x = currentLeft; x <= currentRight; x++)
            {
                canvas[currentTop, x] = borderChar;
                canvas[currentBottom, x] = borderChar;
            }

            for (int y = currentTop; y <= currentBottom; y++)
            {
                canvas[y, currentLeft] = borderChar;
                canvas[y, currentRight] = borderChar;
            }
        }
    }
```

## Прямоугольник ##

```
static void PrintRectangleV1(int width, int height, char borderChar = '#')
{
    for (int y = 0; y < height; y++)
    {
        for (int x = 0; x < width; x++)
        {
            bool border = y == 0 || y == height - 1 || x == 0 || x == width - 1;
            Console.Write(border ? borderChar : ' ');
        }
        Console.WriteLine();
    }
}

PrintRectangleV1(20, 5, '+');
```

```
static string[] GetRectangleV2(int width, int height, char borderChar = '#')
{
    char[,] canvas = CreateCanvas(width, height);

    for (int y = 0; y < height; y++)
    {
        for (int x = 0; x < width; x++)
        {
            bool border = y == 0 || y == height - 1 || x == 0 || x == width - 1;
            if (border)
            {
                canvas[y, x] = borderChar;
            }
        }
    }

    return ToLines(canvas);
}

Console.WriteLine(string.Join(Environment.NewLine , GetRectangleV2(20, 5, '+')));
Console.WriteLine(string.Join(Environment.NewLine , GetRectangleV2(20, 10, '0')));
Console.WriteLine(string.Join(Environment.NewLine , GetRectangleV2(20, 20, '-')));
```

```
static string[] GetRectangle(
    int width,
    int height,
    char borderChar = '#',
    bool isFilled = false,
    char fillChar = '.',
    int borderThickness = 1)
{
    if (width <= 0 || height <= 0)
    {
        return Array.Empty<string>();
    }
    if (borderThickness < 1)
    {
        borderThickness = 1;
    }
    char[,] canvas = CreateCanvas(width, height);
    
    if (isFilled)
    {
        for (int y = borderThickness; y < height - borderThickness; y++)
        {
            for (int x = borderThickness; x < width - borderThickness; x++)
            {
                canvas[y, x] = fillChar;
            }
        }
    }

    DrawRectangleBorder(canvas, 0, 0, width, height, borderChar, borderThickness);
    return ToLines(canvas);
}

Console.WriteLine(string.Join(Environment.NewLine , GetRectangle(20, 5, 'w', true, '-',1)));
```

## Равнобедренный треугольник ##

```
static void PrintRightTriangleV1(int size, char borderChar = '#')
{
    for (int y = 0; y < size; y++)
    {
        for (int x = 0; x <= y; x++)
        {
            bool border = x == 0 || x == y || y == size - 1;
            Console.Write(border ? borderChar : ' ');
        }
        Console.WriteLine();
    }
}
PrintRightTriangleV1(10);
```

```
static string[] GetRightTriangleV2(int size, char borderChar = '#')
{
    char[,] canvas = CreateCanvas(size, size);

    for (int y = 0; y < size; y++)
    {
        for (int x = 0; x <= y; x++)
        {
            bool border = x == 0 || x == y || y == size - 1;
            if (border)
            {
                canvas[y, x] = borderChar;
            }
        }
    }

    return ToLines(canvas);
}
Console.WriteLine(string.Join(Environment.NewLine , GetRightTriangleV2(3,'+')));
```

```
static string[] GetRightTriangle(
    int size,
    char borderChar = '#',
    bool isFilled = false,
    char fillChar = '.',
    bool isMirrored = false)
{
    if (size <= 0)
    {
        return Array.Empty<string>();
    }

    char[,] canvas = CreateCanvas(size, size);

    if (isFilled)
    {
        for (int y = 0; y < size; y++)
        {
            for (int x = 0; x <= y; x++)
            {
                int drawX = isMirrored ? size - 1 - x : x;
                canvas[y, drawX] = fillChar;
            }
        }
    }

    for (int y = 0; y < size; y++)
    {
        for (int x = 0; x <= y; x++)
        {
            bool border = x == 0 || x == y || y == size - 1;
            int drawX = isMirrored ? size - 1 - x : x;

            if (border)
            {
                canvas[y, drawX] = borderChar;
            }
        }
    }

    return ToLines(canvas);
}
Console.WriteLine(string.Join(Environment.NewLine , GetRightTriangle(10,'+', false, '.', true)));
```

## Песочные часы ##

```
static string[] GetHourglass(
    int size,
    char borderChar = '#',
    bool isFilled = false,
    char fillChar = '.')
{
    if (size <= 0)
    {
        return Array.Empty<string>();
    }

    int canvasSize = size * 2 - 1;
    char[,] canvas = CreateCanvas(canvasSize, canvasSize);

    for (int y = 0; y < canvasSize; y++)
    {
        int offset = Math.Min(y, canvasSize - 1 - y);
        int left = offset;
        int right = canvasSize - 1 - offset;

        for (int x = left; x <= right; x++)
        {
            bool border = y == 0 || y == canvasSize - 1 || x == left || x == right;

            if (isFilled && !border)
            {
                canvas[y, x] = fillChar;
            }

            if (border)
            {
                canvas[y, x] = borderChar;
            }
        }
    }

    return ToLines(canvas);
}

Console.WriteLine(string.Join(Environment.NewLine , GetHourglass(6,'+', true, '.')));
```

## Ромб ##

```
static string[] GetRhombus(
    int size,
    char borderChar = '#',
    bool isFilled = false,
    char fillChar = '.')
{
    if (size <= 0)
    {
        return Array.Empty<string>();
    }

    int canvasSize = size * 2 - 1;
    int center = size - 1;
    char[,] canvas = CreateCanvas(canvasSize, canvasSize);

    for (int y = 0; y < canvasSize; y++)
    {
        for (int x = 0; x < canvasSize; x++)
        {
            int distance = Math.Abs(x - center) + Math.Abs(y - center);

            if (distance <= size - 1)
            {
                if (isFilled)
                {
                    canvas[y, x] = fillChar;
                }

                if (distance == size - 1)
                {
                    canvas[y, x] = borderChar;
                }
            }
        }
    }

    return ToLines(canvas);
}

Console.WriteLine(string.Join(Environment.NewLine , GetRhombus(6,'+', true, '.')));
```

## Зебра ##

```
static string[] GetZebra(
    int width,
    int height,
    char stripeChar = '#',
    char backgroundChar = ' ',
    int stripeThickness = 1)
{
    if (width <= 0 || height <= 0)
    {
        return Array.Empty<string>();
    }

    if (stripeThickness < 1)
    {
        stripeThickness = 1;
    }

    char[,] canvas = CreateCanvas(width, height, backgroundChar);

    for (int y = 0; y < height; y++)
    {
        bool stripe = (y / stripeThickness) % 2 == 0;

        if (stripe)
        {
            for (int x = 0; x < width; x++)
            {
                canvas[y, x] = stripeChar;
            }
        }
    }

    return ToLines(canvas);
}

Console.WriteLine(string.Join(Environment.NewLine , GetZebra(6, 10, '+', '.', 1)));
```

## Змейка ##

```
static string[] GetSnake(
    int width,
    int height,
    SnakeDirection direction = SnakeDirection.Clockwise,
    SnakeOrientation orientation = SnakeOrientation.Horizontal,
    char snakeChar = '#',
    char emptyChar = ' ')
{
    if (width <= 0 || height <= 0)
    {
        return Array.Empty<string>();
    }

    if (orientation == SnakeOrientation.Vertical)
    {
        string[] horizontalSnake = GetSnake(
            height,
            width,
            direction,
            SnakeOrientation.Horizontal,
            snakeChar,
            emptyChar);

        return TransposeFigure(horizontalSnake);
    }

    char[,] canvas = CreateCanvas(width, height, emptyChar);

    int bandIndex = 0;

    for (int y = 0; y < height; y += 2)
    {
        bool leftToRight = direction == SnakeDirection.Clockwise
            ? bandIndex % 2 == 0
            : bandIndex % 2 != 0;

        if (leftToRight)
        {
            for (int x = 0; x < width; x++)
            {
                canvas[y, x] = snakeChar;
            }

            if (y + 1 < height)
            {
                canvas[y + 1, width - 1] = snakeChar;
            }
        }
        else
        {
            for (int x = width - 1; x >= 0; x--)
            {
                canvas[y, x] = snakeChar;
            }

            if (y + 1 < height)
            {
                canvas[y + 1, 0] = snakeChar;
            }
        }

        bandIndex++;
    }

    return ToLines(canvas);
}

Console.WriteLine(string.Join(Environment.NewLine , GetSnake(18, 10, SnakeDirection.Clockwise, SnakeOrientation.Horizontal, '#', ' ')));
Console.WriteLine(string.Join(Environment.NewLine , GetSnake(18, 10, SnakeDirection.CounterClockwise, SnakeOrientation.Horizontal, '+', ' ')));
Console.WriteLine(string.Join(Environment.NewLine , GetSnake(12, 18, SnakeDirection.Clockwise, SnakeOrientation.Vertical, '+', ' ')));
```
