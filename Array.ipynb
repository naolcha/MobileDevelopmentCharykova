# Массив(Array)

```
string[] cites = {"Тула","Брянск","Курск","Калуга"}; 
Console.WriteLine(cites[0]); // первый
Console.WriteLine(cites[^1]); // последний
Console.WriteLine(cites.Length);
cites
```

```
Console.WriteLine(Array.IndexOf(cites, "Курск"));
Console.WriteLine(Array.IndexOf(cites, "Курск2"));
Console.WriteLine(Array.LastIndexOf(cites, "Курск"));
```

```
Index index1 = 0;
Index index2 = ^1;

Console.WriteLine(cites[index1]); 
Console.WriteLine(cites[index2]); 

Range range1 = 1..3;

Console.WriteLine(String.Join(", ", cites));
Console.WriteLine(String.Join(", ", cites[1..2]));
Console.WriteLine(String.Join(", ", cites[range1]));
Console.WriteLine(String.Join(", ", cites[..^1]));
```

```
int[] a1 = [1, 2, 3];
int[] a2 = [4, 5, 6];
int[] a3 = [7, 8, 9];
int[] a = [..a1, ..a2, ..a3];
a
```

```
// Q: Как сгенирировать последовательность чисел?

int[] numbers = new int[10];

for (int i = 0; i < 10; i++)
    numbers[i] = i;
numbers
```

```
// ... или
var numbers = Enumerable.Range(start:0, count:10).ToArray();
numbers
```
